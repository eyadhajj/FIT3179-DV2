# Finding out how many people have have heart disease per 100.

```python
import pandas as pd
```

```python
df = pd.read_csv('heart_disease.csv')
df.head()
```

```python
percentage_heart_disease = (df['Heart Disease Status'].value_counts(normalize=True).get('Yes', 0)) * 100
print(f"Percentage of entries with heart disease: {percentage_heart_disease:.2f}%")
```
